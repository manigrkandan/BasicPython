```python
#Function to check if it is odd or even

def OddEven(number):
    if(number%2 == 0):
        print(f"{number} is even number.")
    else:
        print(f"{number} is odd number.")

number = int(input('Enter a number:'))
OddEven(number)
    
```

```python
def percentage10(marklist):
    total = 0
    num_subjects = len(marklist)
    for i, mark in enumerate(marklist):
        print(f"Subject {i+1} = {mark}")
        total += mark
    max_marks = num_subjects * 100
    percentage = (total/max_marks) * 100
    print(f"Total marks: {total}")
    print(f"Percentage: {percentage}")
```

```python
marklist = [89,93,95,85,97]
   
tenthmarkdetails = percentage10(marklist)
```

```python
def marriageEligibility(age,gender):
    print(f"Your Age: {age}")
    print(f"Your Gender: {gender}")
    if age > 21:
        print('You are eligible for marriage')
    else:
        print('You are not eligible for marriage')

inputAge = int(input('Enter your age '))
inputGender = input('Enter your gender ')
marriageEligibility(inputAge,inputGender)
```

```python
#area and perimeter of the triangle

def triangleArea(height,breadth):
    area = height * breadth * 0.5
    return area

def trianglePerimeter(height1,height2,breadth):
    perimeter = height1 + height2 + breadth
    return perimeter

height = 10
height1 = 11
height2 = 22
breadth = 20
triangleArea = triangleArea(height,breadth)
print('Area of triangle is ', triangleArea)

trianglePerimeter = trianglePerimeter(height1, height2, breadth)
print('Perimeter of triangle is ', trianglePerimeter)
```
