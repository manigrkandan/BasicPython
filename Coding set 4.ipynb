```python
#3. Find All Armstrong Numbers Between 1 and 1000
Question: Write a Python program to print all Armstrong numbers between 1 and 1000.
```

```python
armstrong_number =[]
for i in range(1,1001):
    num = [int(j) for j in str(i)]
    power = len(num)
    if sum(j**power for j in num) == i:
        armstrong_number.append(i)

print('Armstrong numbers are ', armstrong_number)
```

```python
#4.Count the Frequency of Each Character in a String
Question: Write a Python program to count the frequency of each character in "programming".

```

```python
string1 = "programming"
frequency = {}

for char in string1:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

for char, count in frequency.items():
    print(f"'{char}' - {count}")
```
